El codigo tranforma una matriz a un dataframe bidireccional (A-B y B-A) con los saltos que hay entre comarcas.
Los valores que tienen NaN significan que no hay conexi√≥n entre esas dos comarcas.

```python
import numpy as np
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import statistics
import seaborn as sns
from pathlib import Path
import re
import requests
from io import StringIO
import random
import pdb
```

```python
matriz_lurraldebus = pd.read_excel("../input_data/public_transport_cost.xlsx", sheet_name='Lurraldebus_saltos')
matriz_euskotren = pd.read_excel("../input_data/public_transport_cost.xlsx", sheet_name='Euskotren_saltos')
matriz_renfe = pd.read_excel("../input_data/public_transport_cost.xlsx", sheet_name='Renfe_saltos')
```

```python
# List with zone names. Regions for Lurraldebus and towns for Euskotren and Renfe

zonas_lurraldebus = matriz_lurraldebus.columns.tolist()
zonas_lurraldebus.remove('Unnamed: 0')

zonas_euskotren = matriz_euskotren.columns.tolist()
zonas_euskotren.remove('Unnamed: 0')

zonas_renfe = matriz_renfe.columns.tolist()
zonas_renfe.remove('Unnamed: 0')
```

```python
# Extract numeric elements in a list of lists
matriz_numeros_lurraldebus = matriz_lurraldebus.iloc[:, 1:].values.tolist()
matriz_numeros_euskotren = matriz_euskotren.iloc[:, 1:].values.tolist()
matriz_numeros_renfe = matriz_renfe.iloc[:, 1:].values.tolist()
```

```python
import pandas as pd

# LURRALDEBUS

# Create lists to save distance between towns
combinaciones_pueblos_lurraldebus = []
distancias_lurraldebus = []

# Iterate over rows and columns of matriz_distancias
for i in range(len(zonas_lurraldebus)):
    for j in range(len(zonas_lurraldebus)):
        #if i != j:
            combinaciones_pueblos_lurraldebus.append((zonas_lurraldebus[i], zonas_lurraldebus[j]))
            distancias_lurraldebus.append(matriz_numeros_lurraldebus[i][j])

# Create a df from data lists
df_combinaciones = pd.DataFrame({'COMARCA ORI': [x[0] for x in combinaciones_pueblos_lurraldebus],
                                 'COMARCA DES': [x[1] for x in combinaciones_pueblos_lurraldebus],
                                 'Saltos': distancias_lurraldebus})

saltos_lurraldebus = pd.DataFrame(df_combinaciones)


# EUSKOTREN

combinaciones_pueblos_euskotren = []
distancias_euskotren = []

for i in range(len(zonas_euskotren)):
    for j in range(len(zonas_euskotren)):
        #if i != j:
            combinaciones_pueblos_euskotren.append((zonas_euskotren[i], zonas_euskotren[j]))
            distancias_euskotren.append(matriz_numeros_euskotren[i][j])

df_combinaciones = pd.DataFrame({'Poblacion ORIGEN': [x[0] for x in combinaciones_pueblos_euskotren],
                                 'Poblacion DESTINO': [x[1] for x in combinaciones_pueblos_euskotren],
                                 'Saltos': distancias_euskotren})

saltos_euskotren = pd.DataFrame(df_combinaciones)


# RENFE

combinaciones_pueblos_renfe = []
distancias_renfe = []

for i in range(len(zonas_renfe)):
    for j in range(len(zonas_renfe)):
        #if i != j:
            combinaciones_pueblos_renfe.append((zonas_renfe[i], zonas_renfe[j]))
            distancias_renfe.append(matriz_numeros_renfe[i][j])

df_combinaciones = pd.DataFrame({'Poblacion ORIGEN': [x[0] for x in combinaciones_pueblos_renfe],
                                 'Poblacion DESTINO': [x[1] for x in combinaciones_pueblos_renfe],
                                 'Saltos': distancias_renfe})

saltos_renfe = pd.DataFrame(df_combinaciones)
```

```python
with pd.ExcelWriter("../input_data/pt_zonal_jumps.xlsx") as writer:
    saltos_lurraldebus.to_excel(writer, sheet_name='lurraldebus')
    saltos_euskotren.to_excel(writer, sheet_name='euskotren')
    saltos_renfe.to_excel(writer, sheet_name='renfe')
```
